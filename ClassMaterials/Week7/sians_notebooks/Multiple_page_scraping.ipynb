# set up urls for each page 

```python
#iterator - based on 1,51,101,151 etc 
iterations=range(1,2250,50)
```

```python
# create url list based on iterations 
for i in iterations:
    start_at=str(i)
    url="https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1995-01-01&user_rating=7.0,&start="+start_at+"&ref_=adv_nxt"
    print(url)
```

# create sleep process

```python
for i in range(5):
    print(i)
    sleep(3) # sleep for 3 seconds 
```

```python
for i in range(5):
    print(i)
    wait_time=randint(1,6)
    print('now I will sleep for '+str(wait_time)+' seconds, goodnight')
    sleep(wait_time)
```

# now we get our response.content from the website (ie scrape all pages)

```python
from tqdm.notebook import tqdm
```

```python
pages=[]
for i in tqdm(iterations):
    start_at=str(i)
    url="https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1995-01-01&user_rating=7.0,&start="+start_at+"&ref_=adv_nxt"
    response=requests.get(url) # get the page
    print('status_code : '+str(response.status_code)) #get the status code
    pages.append(response.content) # collect the page html in a list 
    #have a nap 
    wait_time=randint(1,4)
    print('now i sleep for '+str(wait_time))
    sleep(wait_time)
```
